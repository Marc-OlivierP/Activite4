<h2>Filtre passe-bas</h2>

<p></p>

```python
import IPython.display as ipd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sampleRate = 22050
duration = 6.0

frequence1=98
frequence2=1567.98

t = np.linspace(0, duration, int(duration*sampleRate), endpoint=False) 
sound = 0.5*np.sin(2*np.pi*frequence1*t)+0.5*np.sin(2*np.pi*frequence2*t)

plt.plot(t,sound)
plt.axis([0,0.1,-1,1])
plt.show()
```

```python
ipd.Audio(sound, rate=sampleRate) # load a NumPy array
```

```python
#FFT

amplitudes = np.fft.rfft(sound)
frequencies = np.fft.fftfreq(sound.size, d=1/sampleRate)

frequenciesCut = frequencies[:len(frequencies)//2+1]

plt.plot(frequenciesCut,np.abs(amplitudes)*2/(sampleRate*duration))
plt.axis([0,1100,-0.1,1])


#Low pass filter
R=200
C=32*10**(-6)
f0=1/(2*np.pi*R*C)
print("f0= ",f0)

filteredAmplitudes = amplitudes * (1+(2*np.pi*frequenciesCut*R*C)**2)**(-1/2)
plt.plot(frequenciesCut,np.abs(filteredAmplitudes)*2/(sampleRate*duration))
plt.show()

filteredSound = np.fft.irfft(filteredAmplitudes)


plt.plot(t,filteredSound)
plt.axis([0,0.1,-1,1])
plt.show()
```

```python
ipd.Audio(filteredSound, rate=sampleRate)
```

```python
#high pass filter

#FFT

amplitudes = np.fft.rfft(sound)
frequencies = np.fft.fftfreq(sound.size, d=1/sampleRate)

frequenciesCut = frequencies[:len(frequencies)//2+1]

plt.plot(frequenciesCut,np.abs(amplitudes)*2/(sampleRate*duration))
plt.axis([0,2100,-0.1,1])


#high pass filter
R=2
C=32*10**(-6)
f0=1/(2*np.pi*R*C)
print("f0= ",f0)

filteredAmplitudes[0] = 0
filteredAmplitudes[1:] = amplitudes[1:] * (1+(2*np.pi*frequenciesCut[1:]*R*C)**(-2))**(-1/2)
plt.plot(frequenciesCut,np.abs(filteredAmplitudes)*2/(sampleRate*duration))
plt.show()

filteredSound = np.fft.irfft(filteredAmplitudes)


plt.plot(t,filteredSound)
plt.axis([0,0.01,-1,1])
plt.show()
```

```python
ipd.Audio(filteredSound, rate=sampleRate)
```
